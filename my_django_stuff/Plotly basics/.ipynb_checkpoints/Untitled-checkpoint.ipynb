```python
import pandas as pd
```

```python
df = pd.read_csv('SourceData/nst-est2017-alldata.csv')
```

```python
df['DIVISION'].value_counts()
```

```python
df2= df[df['DIVISION']=='1']
```

```python
df2.set_index('NAME',inplace=True)
```

```python
df2
```

```python
list_of_pop_columns = [col for col in df2.columns if col.startswith('POP')]
```

```python
list_of_pop_columns
```

```python
df2=df2[list_of_pop_columns]
```

```python
df2
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
data = [go.Scatter(x = df2.columns, y = df2.loc[name], mode='lines+markers',name=name) for name in df2.index]
fig = go.Figure(data= data)
pyo.plot(fig)
```
